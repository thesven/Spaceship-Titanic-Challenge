```python
raw_train_data = pd.read_csv('kaggle/input/spaceship-titanic/train.csv')
raw_train_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_categorical_data(df):
    label_encoders = {}

    for column in ["PassengerId", "HomePlanet", "CryoSleep", "Cabin", "Destination", "VIP", "Name"]:
        le = LabelEncoder()
        df[column] = le.fit_transform(df[column])
        label_encoders[column] = le
```

```python
encode_categorical_data(raw_train_data)
raw_train_data.info()
```

```python
X, y = raw_train_data.drop('Transported', axis=1), raw_train_data[['Transported']]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1121218)
```

```python
X_train.info()
```

```python
import xgboost as xgb

xgb_cl = xgb.XGBClassifier()
print(type(xgb_cl))
```

```python
from sklearn.metrics import accuracy_score

xgb_cl.fit(X_train, y_train)
preds = xgb_cl.predict(X_test)
adjusted_preds = preds.astype(bool)
accuracy_score(y_test, preds)
```

# Attempt to tune the parameters for better results using gridsearch

```python
param_grid = {
    "max_depth": [9, 15, 20],
    "learning_rate": [0.1, 0.01, 0.05],
    "gamma": [0, 0.25, 1],
    "reg_lambda": [0, 1, 10],
    "scale_pos_weight": [1, 3, 5],
    "subsample": [0.8],
    "colsample_bytree": [0.5],
}
```

```python
from sklearn.model_selection import GridSearchCV

# Init classifier
xgb_cl = xgb.XGBClassifier(objective="binary:logistic", )

# Init Grid Search
grid_cv = GridSearchCV(xgb_cl, param_grid, n_jobs=-1, cv=3, scoring="roc_auc", verbose=1)

# Fit
_ = grid_cv.fit(X_train, y_train)

grid_cv.best_score_
```

```python
grid_cv.best_params_
```

```python
final_cl = xgb.XGBClassifier(
    **grid_cv.best_params_,
    objective="binary:logistic",
)

_ = final_cl.fit(X_train, y_train)
preds = final_cl.predict(X_test)
adjusted_preds = preds.astype(bool)
accuracy_score(y_test, preds)
```

# Using the optimized XGBClassifier, run the test set for predictions

```python
preds = final_cl.predict(raw_test_data)
adjusted_preds = preds.astyle(bool)

final_results = pd.DataFrame()
final_results['PassengerId'] = raw_test_data['PassengerId']
final_results['Transported'] = adjusted_preds

final_results.to_csv("submissions.csv")

final_results.head()
```
